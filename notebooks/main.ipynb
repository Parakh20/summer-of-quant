```python
import sys
from pathlib import Path

# Walk up from the current working directory until we find the repo root
# (identified by the presence of the strategy/ package), so this works
# whether Jupyter was launched from the repo root or from notebooks/.
REPO_ROOT = Path.cwd()
while not (REPO_ROOT / "strategy").exists() and REPO_ROOT != REPO_ROOT.parent:
    REPO_ROOT = REPO_ROOT.parent
sys.path.insert(0, str(REPO_ROOT))

import pandas as pd

from backtest.metrics import compute_all_metrics, plot_equity_curve, plot_monthly_returns_heatmap
from backtest.runner import run_backtest
from data.loader import load_training_data
from strategy.params import StrategyParams

# Load the official competition training data (BTC_2019_2023_1d.csv), not the
# 2014-2019 Yahoo-style file the original version of this notebook used.
df = load_training_data()
print(f"Loaded {len(df)} rows: {df['datetime'].min()} to {df['datetime'].max()}")
df.tail()
```

```python
params = StrategyParams()
bt, signals_df = run_backtest(df, params)
print(f"Total trades: {len(bt.trades)}")

stats = bt.get_statistics()
for key, val in (stats or {}).items():
    print(f"{key:.<35}: {val}")
```

```python
metrics = compute_all_metrics(bt)
for key, val in metrics.items():
    print(f"{key:.<35}: {val}")
```

```python
# Static plots, saved alongside the interactive ones below
plot_equity_curve(bt, REPO_ROOT / "results" / "equity_curve.png")
plot_monthly_returns_heatmap(bt, REPO_ROOT / "results" / "monthly_returns_heatmap.png")
print("Saved equity_curve.png and monthly_returns_heatmap.png to results/")
```
